# File Input and Output using C

### Considerations


- **You must validate the buffer size, that avoid a segmentation fault**

- **Saving the struct data in binary form give us a minor size in bytes number and allow a better read and write performance**

- **If you want read the binary file without the struct data, you will need a Decoder that using inverse engineering allow read the content of this binary file**
